## Классические вероятности. Решение аналитическими методами и методами имитационного моделирования на ЭВМ

Определить вероятность того, что при шести бросаниях кости выпадут все разные цифры. (Кость - кубик с нанесенными на его грани цифрами от 1 до 6)

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import math
from random import random

def roll_dice():
    return math.ceil(random() * 6)

n = 100000
success = 0
for i in range(n):
    rolls = [0 for j in range(6)]
    for j in range(6):
        rolls[roll_dice() - 1] += 1
    all_six = True
    for d in rolls:
        all_six = all_six and d == 1
    if all_six:
        success += 1

precise_sol = (5)/(9*6*6)
print("      точное решение: ", precise_sol)
print("моделируемое решение: ", success / n)
print("              ошибка: ", success / n - precise_sol)
```
