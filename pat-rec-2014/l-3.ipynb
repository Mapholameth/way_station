```python
%matplotlib inline
import math
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
sns.set(style="dark")
```

# Лабораторная Работа 3

```python
k = 3
n = 1000
mean = [0 for i in range(k)]
d = [1, 1, 9]

cov = []
for i in range(k):
    row = [0 for i in range(k)]
    row[i] = d[i]
    cov.append(row)
    
cov = [
    [1, 0, -0.2],
    [0, 1/9, 0],
    [-0.2, 0, 1/9]
]

xlim = (-5, 5)
ylim = xlim

df = np.random.multivariate_normal(mean, cov, size=n)
df = pd.DataFrame(df, columns=[['x{}'.format(i) for i in range(k)]], index=range(n))
```

```python
color = sns.color_palette()[2]
mpl.rc("figure", figsize=(6, 6))
g = sns.jointplot('x0', 'x2', data=df, kind="reg", color=color, size=8, xlim=xlim, ylim=ylim)
```

```python
print("длина вектора разницы оценки среднего и среднего ", np.linalg.norm(np.mean(df) - mean))
```

```python
cov_diff = np.cov(df, rowvar=0) - cov
print("норма разницы ковариационной матрицы и оценки ковариационной матрицы: ", np.linalg.norm(cov_diff))

np.cov(df, rowvar=0)
```

```python
g = sns.PairGrid(df, diag_sharey=False, size=8)
g.map_lower(sns.kdeplot)
g.map_upper(plt.scatter, alpha=1)
g.map_diag(sns.kdeplot, lw=3)
g.set(xlim=xlim, ylim=ylim)
```
