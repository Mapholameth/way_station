### Задача 6. Постановка Задачи

Некий студент решил испечь 10 булочек с изюмом и угостить своих товарищей. Для этого замесил тесто, бросил в него 50 изюминок, все тщательно размял, разделил на десять равных частей, положил в печь, испек, принес в институт и угостил своих друзей. Случайная величина X — число изюминок в булочке, которая досталась студенту Петрову. (примем за допущение, что она ему всё-таки досталась, в процессе угощения друзей)

### Анализ

Вероятность попадения одной изюминки в булочку $= \frac{1}{10} = 0.1$ Так как изюминки попадают в булочки независимо друг от друга, то вероятность попадения $k$ изюминок в булочку подчиняется Биномиальному закону распределения: $p_n(k) = \frac{n!}{(n - k)!k!} p^k q^{n-k}, p = 0.1, n = 50$. По [Теореме Пуассона](https://en.wikipedia.org/wiki/Poisson_limit_theorem) при количестве испытаний, устремлённому к бесконечности $p(k, \lambda) = \frac{\lambda^ke^{-\lambda}}{k!}, \lambda = np = 50 * 0.1 = 5$.

```python
from IPython.core.display import HTML
from IPython.nbconvert.filters.markdown import markdown2html

def poisson(k, l=5):
    return l**k / math.factorial(k) * math.e**(-l)

n = 19
x_n = [i for i in range(n)]
p_n = [poisson(i) for i in x_n]

out_md = "### Построить Ряд Распределения"
out_md += "\n|$x_i$|"
for i in range(n):
    out_md += "{}|".format(x_n[i])
out_md += "\n|---|"
for i in range(n):
    out_md += "---"

out_md += "\n|$p_i$|"
for i in range(n):
    out_md += "{0:.4f}|".format(p_n[i])


HTML(markdown2html(out_md))
```

### Рассчитать Аналитически Среднее и Дисперсию Случайной Величины $X$

$E(X) = \sum_{k=0}^\infty k p(k) = \sum_{k=0}^\infty k \frac{\lambda^ke^{-\lambda}}{k!} = \lambda e^{-\lambda} \sum_{k=0}^\infty \frac{\lambda^{k-1}}{(k-1)!} = \lambda e^{-\lambda}e^\lambda = \lambda$

$V(X) = E(X)$

### Построить и Отобразить Гистограмму Распределения Х

```python
N=10000
rs = np.random.RandomState()
d = rs.poisson(lam=5.0, size=N)
print("Оценка Среднего: ", np.mean(d))
print("Оценка Дисперсии: ", np.var(d))
print("Оценка Коэффициента Асимметрии: ", stats.skew(d))
print("Оценка Коэффициента Эксцесса: ", stats.kurtosis(d))
n = 14
bins_d = -0.5
bins = np.linspace(bins_d, n + bins_d, n, endpoint=False)
plt.hist(d, bins);
print(d)
```

```python
d.sort()
x = np.arange(0, 14, 0.001)
plt.step(x, sum(x > i for i in d) / N)
```

## Часть 2. Непрерывные Случайные Величины

### Задача 1

Случайная величина эксцентриситета детали характеризуется функцией распределения Рэлея

|pdf|$ f(x;\sigma) = \frac{x}{\sigma^2} e^{-x^2/(2\sigma^2)}, \quad x \geq 0,$|
|---|---|---|
|cdf|$ F(x) = 1 - e^{-x^2/(2\sigma^2)}$|
|mode|$ \sigma$|
|median|$ \sigma\sqrt{\ln(4)}\ $|
|mean|$ \sigma \sqrt{\frac{\pi}{2}}$|
|variance|$ \frac{4 - \pi}{2} \sigma^2$|

```python
sigma = [0.5, 1, 2, 3, 4]

for i in range(5):
    s = sigma[i]
    x = np.arange(0, 16, 0.001)
    plt.plot(x, stats.rayleigh.pdf(x, 0, s))
```

```python
sigma = [0.5, 1, 2, 3, 4]

for i in range(5):
    s = sigma[i]
    x = np.arange(0, 16, 0.001)
    plt.plot(x, stats.rayleigh.cdf(x, 0, s))
```

```python
d = np.random.rayleigh(1, 10000)
print("Оценка Среднего: ", np.mean(d), " ошибка: ", np.mean(d) - (math.sqrt(math.pi/2)))
print("Оценка Дисперсии: ", np.var(d), " ошибка: ", np.var(d) - ((4-math.pi) / 2))
print("Оценка Коэффициента Асимметрии: ", stats.skew(d))
print("Оценка Коэффициента Эксцесса: ", stats.kurtosis(d))
sns.distplot(d)
```

```python
n = 20
d = np.random.rayleigh(1, n)
d.sort()
x = np.arange(0, 4, 0.001)
plt.step(x, sum(x > i for i in d) / n)
sum(x > i for i in d) / n
```
