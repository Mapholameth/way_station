```python
def u_0(x, x_0):
    if (x == x_0):
        return 0
    else:
        return 1
    
t = np.arange(10)
f = [u_0(t[i], 2) for i in range(10)]
plt.plot(t, f)
plt.vlines(t, 0, f, alpha=0.5)
```

```python
def u_0(x, x_0):
    if (x < x_0):
        return 0
    else:
        return 1
    
t = np.arange(10)
f = [u_0(t[i], 5) for i in range(10)]
plt.plot(t, f)
plt.vlines(t, 0, f, alpha=0.5)
```

```python
def u_0(x, a):
    return a ** x
    
t = np.arange(10)
f = [u_0(t[i], 0.5) for i in range(10)]
plt.plot(t, f)
plt.vlines(t, 0, f, alpha=0.5)
```

```python
def u_0(x, a, omega, phi):
    return a * np.sin(x * omega + phi)

n = 32
t = np.arange(n)
f = [u_0(t[i], 1, 0.5, 0) for i in range(n)]
plt.plot(t, f)
plt.vlines(t, 0, f, alpha=0.5)
```
